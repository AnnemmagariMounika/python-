### What is an Object?

An object is an instance of a class. It can store data (attributes) and behavior (methods). Objects are the basic building blocks of a Python program, allowing for encapsulation and modularity.

### Key Concepts

*1.Class:*
   - A class is a blueprint for creating objects. It defines a set of attributes and methods that the created objects will have.
   python
   class DataScientist:
    def __init__(self, name, specialty):
        self.name = name
        self.specialty = specialty

    def analyze(self):
        return f"{self.name} is analyzing data!"

# Example usage:
data_scientist = DataScientist("Bharath", "Machine Learning")
print(data_scientist.name)       # Outputs: Bharath
print(data_scientist.specialty)  # Outputs: Machine Learning
print(data_scientist.analyze())  # Outputs: Bharath is analyzing data!

   

2. *Instance:*
   - An instance is a specific object created from a class.

   
  python

         data_scientist = DataScientist("Bharath", "Machine Learning")

   

3. *Attributes:*
   - Attributes are variables that belong to an object. They hold the data related to the object.

   python

4. *Methods:*
   - Methods are functions defined within a class that describe the behaviors of the objects.

### Built-in Types of Objects

Python has several built-in types of objects, including:

1. *Numbers:*
   - Integers, floating-point numbers, complex numbers.
   
   python

            x = 10                # Integer
            y = 3.14              # Float
            z = 1 + 2j            # Complex
   

2. *Strings:*
   - Immutable sequences of characters.
   
   python
         greeting = "Machine, Learning!"
   

3. *Lists:*
   - Mutable sequences of objects.

   python

       fruits = ["AI", "ML", "NLP"]
   

4. *Tuples:*
   - Immutable sequences of objects.

   python

       point = (1, 2)
   

5. *Dictionaries:*
   - Mutable mappings of keys to values.

   python

       person = {"name": "Manvitha", "age": 24}
   

6. *Sets:*
   - Mutable collections of unique objects.

   python

       unique_numbers = {1, 2, 3, 4}
   

7. *Booleans:*
   - Representing truth values.

   python

       is_active = True
       is_closed = False
   

### Creating Custom Objects

You can create custom objects by defining classes and instantiating them:

class Project:

    def __init__(self, name, field, duration):
        self.name = name
        self.field = field
        self.duration = duration

# Creating an instance of Project
my_project = Project("AI Model Deployment", "Machine Learning", 6)

# Accessing attributes

     print(my_project.name)                  # Output: AI Model Deployment
   
     print(my_project.field)                 # Output: Machine Learning
   
     print(my_project.duration)              # Output: 6 months
